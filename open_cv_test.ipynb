```python
for images in range(len(dog_files)):

    image1 = cv2.imread(dog_files[0])
    img_resized1 = cv2.resize(image1, (421,499)) #needed so matchTamplete works
    image_base = cv2.imread(dog_files[images])
    img_resized2 = cv2.resize(image_base, (421,499)) #needed so matchTamplete works
    corr = cv2.matchTemplate(img_resized1, img_resized2,
                             cv2.TM_CCOEFF_NORMED)[0][0]
    if 0.45 <= corr < 1:
        print('found it: ', corr)
        fig, axs = plt.subplots(1,2, figsize = (10,5))
        axs[0].imshow(image1)
        axs[1].imshow(image_base)
        plt.show()
        break


```
